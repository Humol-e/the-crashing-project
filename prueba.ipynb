```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('global_traffic_accidents.csv')

```

```python
print(df.to_string())
```

```python
causas = [
    "Drunk Driving",
    "Reckless Driving",
    "Weather Conditions",
    "Speeding",
    "Mechanical Failure"
]

contador = {}
for causa in df['Cause']:
    if causa not in contador:
        contador[causa] = 0
    contador[causa] += 1

print(contador)

plt.bar(contador.keys(), contador.values(), width=0.5,)
plt.show()
    
    
```
